```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data1=pd.read_csv('/content/drive/MyDrive/Datasets/banana_quality_dataset.csv')
data1.head()
```

```python
data1.info()
```

```python
data1.drop(['sample_id','harvest_date'],axis=1,inplace=True)
```

```python
data1.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
def column_transformer(df):
    num_cols = df.select_dtypes(include="number").columns
    cat_cols = df.select_dtypes(include="object").columns

    return ColumnTransformer(
        transformers=[
            ("num", StandardScaler(), num_cols),
            ("cat", OneHotEncoder(sparse_output=False), cat_cols)
        ]
    )
```

```python
pre = column_transformer(data1)
```

```python
pre=column_transformer(data1)
```

```python



pre
```

```python
new_data1=pre.fit_transform(data1)
```

```python
new_data1
```

```python
new_dataset=pd.DataFrame(new_data1,columns=pre.get_feature_names_out())
```

```python
new_dataset.head()
```

```python
data2= pd.read_csv('/content/drive/MyDrive/Datasets/diabetes_data.csv')
data2.head()
```

```python
data2.drop('Outcome',axis=1,inplace=True)
```

```python
data2.head()
```

```python
data2.info()
```

```python
data2.isnull().sum()
```

```python
for num in data2.columns:
  if data2[num].mean()==data2[num].median():
    data2[num].fillna(data2[num].mean(),inplace=True)
  else:
    data2[num].fillna(data2[num].median(),inplace=True)
```

```python
data2.isnull().sum()
```

```python
pre = column_transformer(data2)
```

```python
new_data=pre.fit_transform(data2)
```

```python
new_data
```

```python
new_dataset=pd.DataFrame(new_data,columns=pre.get_feature_names_out())
```
