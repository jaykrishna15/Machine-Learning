### Handling Missing values

```python
num_col= data.select_dtypes(include=['int64','float64']).columns
cat_col=data.select_dtypes(include=['object']).columns
print(num_col)
print(cat_col)
```

```python
for num in num_col:
  if data[num].mean()==data[num].median():
    data[num].fillna(data[num].mean(),inplace=True)
  else:
    data[num].fillna(data[num].median(),inplace=True)
```

```python
for cat in cat_col:
  data[cat].fillna(data[cat].mode()[0],inplace=True)
```

```python
data.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
pre= ColumnTransformer(
    transformers=[('num',StandardScaler(),num_col),
                  ('cat',OneHotEncoder(sparse_output=False),cat_col)],remainder='passthrough'
)
```

```python
pre
```

```python
new_data= pre.fit_transform(data)
```
