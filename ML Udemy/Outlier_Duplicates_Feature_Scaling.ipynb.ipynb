# Detecting Outliers
- IQR
- BoxPlot

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.boxplot(data=df,x=df['ph'])
```

```python
sns.boxplot(data=df,x=df['tds'])
```

```python
q1=df['tds'].quantile(0.25)
q3=df['tds'].quantile(0.75)
IQR=q3-q1
median=df['tds'].median()
```

```python
print(q1,q3,IQR,median)
```

```python
lower_fence=q1-1.5*IQR
upper_fence=q3+1.5*IQR
```

```python
lower_fence
```

```python
upper_fence
```

```python
df.loc[(df['tds'] < lower_fence) | (df['tds'] > upper_fence)]
```

```python
print(sum((df['tds'] < lower_fence) | (df['tds'] > upper_fence)))
```

# Handling Outliers

```python
print(sum((df['tds']<lower_fence)|(df['tds']>upper_fence)))
```

```python
# df=df.loc[(df['tds']<lower_fence)|(df['tds']>upper_fence)]
```

```python
#df
```

```python
df=df[(df['tds']>=lower_fence) & (df['tds'] <= upper_fence)]
```

```python
df
```

```python
sns.boxplot(data=df,x=df['tds'])
```

```python
df['tds']=np.where(df['tds']>upper_fence,upper_fence,df['tds'])
```

```python
sns.boxplot(data=df,x=df['tds'])
```

```python
from feature_engine.outliers import Winsorizer
```

```python
q1=df['sulfate'].quantile(0.25)
q3=df['sulfate'].quantile(0.75)
IQR=q3-q1
median=df['sulfate'].median()
```

```python
print(q1,q3,IQR,median)
```

```python
lower_fence=q1-1.5*IQR
upper_fence=q3+1.5*IQR
```

```python
lower_fence
```

```python
upper_fence
```

```python
from feature_engine.outliers import Winsorizer

winsor = Winsorizer(capping_method='iqr', tail='both', fold=1.5)
df['sulfate'] = winsor.fit_transform(df[['sulfate']])
```

```python
sns.boxplot(data=df,x=df['sulfate'])
```
